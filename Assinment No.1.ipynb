### 58. Electricity Bill: Pehle 100 units free, next 100 units Rs 5/unit, uske baad Rs 10/unit.

```python
units = int(input("Enter electricity units: "))

if units <= 100:
    bill = 0

elif units <= 200:
    bill = (units - 100) * 5

else:
    bill = (100 * 5) + ((units - 200) * 10)

print("Electricity Bill =", bill)
```
